```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.
```

```python
import numpy as np
import matplotlib.pyplot as plt

seed = 2143
rng = np.random.default_rng(seed)

n = 200
t = rng.uniform(0, 10, n)

#independent Gaussian sensor readings

sensor_a = rng.normal(loc=25.0, scale=3.0, size=n)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n)

#sort by time so readings are time-ordered

order = np.argsort(t)
t = t[order]
sensor_a = sensor_a[order]
sensor_b = sensor_b[order]

plt.figure(figsize=(9,4))
plt.plot(t, sensor_a, '-o', ms=4, label='Sensor A (μ=25, σ=3)')
plt.plot(t, sensor_b, '-s', ms=4, label='Sensor B (μ=27, σ=4.5)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)') 
plt.title('Simulated temperature readings (seed=2143)')
plt.legend() 
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Box plot comparing Sensor A and Sensor B temperature distributions.
import numpy as np
import matplotlib.pyplot as plt

# Ensure data exists; generate using the known seed if necessary
if not all(name in globals() for name in ('sensor_a','sensor_b')):
    seed = 2143
    rng = np.random.default_rng(seed)
    n = 200
    sensor_a = rng.normal(25.0, 3.0, size=n).astype(np.float64)
    sensor_b = rng.normal(27.0, 4.5, size=n).astype(np.float64)

plt.figure(figsize=(6,5))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A','Sensor B'], notch=True, showmeans=True,
            meanprops=dict(marker='D', markeredgecolor='black', markerfacecolor='firebrick'))
plt.ylabel('Temperature (°C)')
plt.title('Box plot: Sensor A vs Sensor B')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
